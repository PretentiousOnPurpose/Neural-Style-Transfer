```python
import tensorflow as tf
from keras.applications import VGG16
from PIL import Image
import numpy as np
import keras.backend as K
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
height = 300; width = 400
```

```python
c = np.array(Image.open("c.jpg").resize((width, height))).reshape(1, height, width, 3)
s = np.array(Image.open("s.jpg").resize((width, height))).reshape(1, height, width, 3)
```

```python
plt.imshow(c[0])
```

```python
plt.imshow(s[0])
```

```python
c = c[:, :, :, ::-1]
s = s[:, :, :, ::-1]
```

```python
C = tf.constant(value = c, dtype = tf.float32)
S = tf.constant(value = s, dtype = tf.float32)
W = tf.Variable(initial_value= tf.random_normal(shape = (1, height, width, 3)))
```

```python
Inp = tf.concat(values = [C, S, W], axis = 0)
```

```python
model = VGG16(include_top=False, input_tensor= Inp)
```

```python
[layer.name for layer in model.layers]
```

```python
Loss = 0.0
```

```python
# content_samples = ['block4_conv1']
```

```python
def content_loss():
    Out = model.get_layer(index = 11).output
    return tf.reduce_mean(tf.square(Out[0] - Out[2]))/(4 * height * width * 3)
```

```python
Loss += content_loss()
```

```python
# This function is taken from fchollet's implementation
def gram_matrix(x):
    features = K.batch_flatten(K.permute_dimensions(x, (2, 0, 1)))
    gram = K.dot(features, K.transpose(features))
    return gram
```

```python
def style_loss():
    Out1 = model.get_layer(index = 13).output
    Out2 = model.get_layer(index = 15).output
    Out3 = model.get_layer(index = 17).output
    Out = [Out1, Out2, Out3]
    style = 0.0
    for i in range(3):
        style += tf.reduce_mean(tf.square(gram_matrix(Out[i][1]) - gram_matrix(Out[i][2])))/((3 * height * width)**2)
    
    return style
```

```python
Loss += style_loss()
```

```python
opt = tf.train.AdamOptimizer(learning_rate = 0.005)
```

```python
sess = tf.InteractiveSession()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
def train(epochs = 100):
    X = np.random.normal(0, 1, size = (1, height, width, 3))
    for e in range(epochs):
        loss = sess.run(Loss, feed_dict={:X})
        opt.apply_gradients(grads_and_vars= opt.compute_gradients(loss, var_list=[W]))
    return X
```

```python
train(epochs = 1)
```
